Comparison between DBSCAN and K-Means.

## DBSCAN and Clustering Examples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
X, y = datasets.make_blobs(n_samples=1000, centers=3, n_features=2,random_state=0)
blobs = pd.DataFrame(X, columns = ['X1','X2'])
```

```python
blobs.head()
```

```python
sns.scatterplot(data=blobs,x='X1',y='X2')
```

```python
moons = datasets.make_moons(n_samples=1000, noise=0, random_state=0)
moons = pd.DataFrame(moons[0], columns = ['X1','X2'])
```

```python
moons.head()
```

```python
sns.scatterplot(data=moons,x='X1',y='X2')
```

```python
circles = datasets.make_circles(n_samples=1000, factor=0.01, noise=0.05, random_state=0)
circles = pd.DataFrame(circles[0], columns = ["X1","X2"])
```

```python
circles.head()
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

## Label Discovery

```python
def display_categories(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

## Kmeans Results

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2)
```

```python
display_categories(model,moons)
```

```python
model = KMeans(n_clusters = 3)
display_categories(model,blobs)
```

```python
model = KMeans(n_clusters = 2)
display_categories(model,circles)
```

## DBSCAN Results

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.3)
```

```python
display_categories(model,blobs)
```

```python
model = DBSCAN(eps=0.15)
plt.figure(figsize=(10,6),dpi=150)
display_categories(model,moons)
```

```python
display_categories(model,circles)
```
