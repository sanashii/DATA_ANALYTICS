```python
def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(X)
```

```python
percent_nan
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90);
```

# Technique 1: Removing Features or Removing Rows.

## Row Wise
### For regression and clustering:
- Check the your data row wise, if there are only a few number of rows with missing values relative to the entire number of rows, then it is suitable to drop these rows.

### For classification:
- Check for class imbalance, if your data has equal number of classes refer to the techniques for regression.

## Column Wise

- If a specific column has too many missing values relative to the entire data, it is preferable to drop these columns.

- In our example a feature column for the number of cars that can fit into a garage. Perhaps if there is no garage then there is a null value, instead of a zero. It probably makes more sense to quickly fill the null values in this case with a zero instead of a null. Only you can decide based off your domain expertise and knowledge of the data set.

# [Imputation](https://en.wikipedia.org/wiki/Imputation_(statistics)) of Missing Data

- Impute missing data based on statistical values: impute to the mean or impute to the mode, for unimodal data.
- Perform complex model based approach, like training a linear regression.

```python
print('Average Price:', X.price.mean())
print('Average Stroke:', X.stroke.mean())
print('Average Bore:', X.bore.mean())
print('Average normalized-losses:', X['normalized-losses'].mean())
```

```python
X['price'] = X['price'].fillna(X.price.mean())
X['stroke'] = X['stroke'].fillna(X.stroke.mean())
X['bore'] = X['bore'].fillna(X.bore.mean())
X['normalized-losses']= X['normalized-losses'].fillna(X['normalized-losses'].mean())
```
